```python
import pandas as pd
import altair as alt
from datetime import timedelta
import numpy as np 

alt.data_transformers.enable("vegafusion")
```

Salmon Tagging Data 

```python
salmon2 = salmon1 = pd.read_csv('data/data_salmon2.csv')
salmon2_cow = salmon2[salmon2['watershed'] == 'cowichan']
# df = pd.DataFrame(salmon2)
salmon2_cow.head() 
```

```python
watershed_dummies = pd.get_dummies(salmon2_cow['watershed'], prefix='watershed')
site_dummies = pd.get_dummies(salmon2_cow['site'], prefix='site')
species_dummies = pd.get_dummies(salmon2_cow['species'], prefix='species')

df_expanded_cow = pd.concat([salmon2_cow, watershed_dummies, site_dummies, species_dummies], axis=1)

df_expanded_cow.drop(['watershed', 'site', 'species'], axis=1, inplace=True)

columns = [col for col in df_expanded_cow.columns if col != 'count'] + ['count']
df_expanded_cow = df_expanded_cow[columns]
df_expanded_cow.head() 
```
